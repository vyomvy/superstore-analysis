```python

g=df.groupby('Category').agg({
    'Profit':'sum',
    'Sales':'sum'}
)
g['margin']=(g['Profit']/g['Sales'])*100
g.sort_values(by='margin',ascending=False)

```

```python
g=df.groupby('Product Name').agg({
    'Profit':'sum',
    'Sales':'sum',
    'Quantity':'sum'}
)
g['Price']=g['Sales']/g['Quantity']
m=g['Price'].mean()
g[g['Price']<m]
g.sort_values(by='Profit',ascending=False)



```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.groupby('City').agg({
    'Order Date':'nunique',
    'Profit':'sum'
}).sort_values(by='Order Date',ascending=False)
```

# Insight :


Observation:\
There is a significant disparity in activity levels across cities, where a few cities contribute a disproportionately high number of orders while others remain largely inactive.

Reason:\
This imbalance may indicate uneven market penetration or demand concentration in specific regions, rather than consistent performance across all cities.

Impact:\
Over-reliance on a few high-performing cities increases business risk and limits overall growth potential.

Recommendation:\
Focus on improving engagement in underperforming cities through targeted marketing, localized strategies, and expansion of operational presence to achieve more balanced growth.


```python
g = df.groupby('City').agg(
    Sales=('Sales', 'sum'),
    Orders=('Order_ID', 'nunique'),
    Quantity=('Quantity', 'sum'),
    Profit_sum=('Profit', 'sum')
)

pp=df[df['Profit'] > 0].groupby('City')['Order_ID'].nunique()
g['AOV'] = g['Sales'] / g['Orders']
g['Profit per Rupee'] = g['Profit_sum'] / g['Sales']
g['Profit per Order'] = g['Profit_sum']/g['Orders']
g['% of Profitable Orders'] = ((pp/ g['Orders'])*100).fillna(0)


g.loc[:,['Profit per Rupee','Profit per Order','AOV','% of Profitable Orders','Quantity']]\
.sort_values(by='Quantity',ascending=False).head(10)
```
